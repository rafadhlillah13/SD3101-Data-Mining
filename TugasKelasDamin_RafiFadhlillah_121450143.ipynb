```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Membuat DataFrame manual
data = pd.DataFrame({
    'NamaMahasiswa': ['Roy', 'Sinta', 'Iqbal', 'Dilan', 'Ratna', 'Merry', 'Rudi', 'Hafiz', 'Gede', 'Christian', 'Justin', 'Jesika', 'Ayu', 'Siska', 'Reitama'],
    'UTS': [89, 90, 70, 45, 65, 80, 90, 70, 96, 60, 45, 60, 85, 52, 40],
    'TUGAS': [90, 71, 75, 65, 75, 70, 85, 70, 93, 55, 60, 70, 90, 68, 60],
    'UAS': [75, 95, 80, 59, 53, 75, 81, 73, 85, 48, 58, 72, 88, 55, 70]
})

# Mengambil kolom UTS, TUGAS, dan UAS sebagai fitur
X = data[['UTS', 'TUGAS', 'UAS']]

# Inisialisasi model K-means dengan 3 kluster
kmeans = KMeans(n_clusters=3, random_state=0)

# Melakukan klustering
data['Cluster'] = kmeans.fit_predict(X)

# Mengganti label kluster
data['Cluster'].replace({0: 'Pintar', 1: 'Kurang', 2: 'Sedang'}, inplace=True)

# Menampilkan hasil klustering
print(data)

# Plot hasil klustering
colors = ['red', 'green', 'blue']
for i in range(3):
    cluster_data = data[data['Cluster'] == ['Pintar', 'Kurang', 'Sedang'][i]]
    plt.scatter(cluster_data['UTS'], cluster_data['TUGAS'], c=colors[i], label=f'{["Pintar", "Kurang", "Sedang"][i]}')

# Menampilkan pusat kluster
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', marker='x', s=200, label='Centroids')
plt.xlabel('UTS')
plt.ylabel('TUGAS')
plt.legend()
plt.title('K-means Clustering of Mahasiswa')
plt.show()
```

Hasil analisis dari output di atas adalah pemisahan mahasiswa menjadi tiga kluster ("Pintar", "Kurang," dan "Sedang") berdasarkan nilai UTS, TUGAS, dan UAS mereka. Berikut adalah beberapa insight yang dapat diperoleh dari hasil klustering:

1. Kluster "Pintar" (Merah):
- Kluster ini berisi mahasiswa dengan kinerja akademik yang tinggi. Mereka memiliki nilai UTS, TUGAS, dan UAS yang relatif tinggi.
- Mahasiswa dalam kluster ini cenderung mencapai hasil yang baik dalam tugas, ujian tengah semester, dan ujian akhir semester.
- Kluster ini mungkin mencakup mahasiswa yang rajin dan memiliki pemahaman yang kuat tentang materi.

2. Kluster "Kurang" (Hijau):
- Kluster ini berisi mahasiswa dengan kinerja akademik yang rendah. Mereka memiliki nilai UTS, TUGAS, dan UAS yang relatif rendah.
- Mahasiswa dalam kluster ini mungkin perlu meningkatkan hasil belajar mereka, terutama dalam hal tugas, ujian tengah semester, dan ujian akhir semester.
- Kluster ini mungkin mencakup mahasiswa yang memerlukan bantuan tambahan dalam studi mereka.

3. Kluster "Sedang" (Biru):

- Kluster ini berisi mahasiswa dengan kinerja akademik yang berada di antara "Pintar" dan "Kurang." Mereka memiliki nilai UTS, TUGAS, dan UAS yang cenderung menengah.
- Mahasiswa dalam kluster ini mungkin konsisten dalam kinerja akademik mereka, tetapi ada ruang untuk perbaikan.
- Kluster ini mencakup mahasiswa yang sebenarnya sudah cukup baik namun dapat berupaya lebih giat untuk mencapai hasil yang lebih baik.
